## Video preprocessing
* Run object detection on video
* Cut out cars 

```python
def pad_to_square(img):
    horizontal = max(0, img.shape[0] - img.shape[1]) // 2
    vertical = max(0, img.shape[1] - img.shape[0]) // 2
    padded = cv2.copyMakeBorder(img, vertical, vertical, horizontal, horizontal, cv2.BORDER_CONSTANT, value=(0, 0, 0))
    scaled = cv2.resize(padded, dsize=(128,128), interpolation = cv2.INTER_LINEAR)
    return scaled


def run_detection(images):
    prep = [ resize_image(preprocess_image(image)) for image in images]
    images, scales = zip(*prep)
    scale = np.unique(scales)[0]
    boxes, scores, _ = detection_model.predict_on_batch(np.array(images))
    boxes = [box[0][score[0] >= 0.5] / scale for box, score in zip(boxes, scores)]
    return boxes


def draw_boxes(image, boxes):
    draw = image.copy()
    draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)
    for box in boxes:
        b = box.astype(int)
        draw_box(draw, b, color=(255,0,0))
    return draw


def highlight_boxes(image, boxes):
    draw = image.copy()
    draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)
    stencil = np.zeros(draw.shape).astype(draw.dtype)
    for box in boxes:
        b = box.astype(int)
        cv2.rectangle(stencil, (b[0], b[1]), (b[2], b[3]), (255,255,255), thickness=cv2.FILLED)
    return cv2.bitwise_and(draw, stencil)
```

```python
%%time

video_np = video_to_npy('../../video/collisions/output000.mp4', width=480, fps=5)
video_boxes = run_detection(video_np)
video_draw = [draw_boxes(v, b) for v, b in zip(video_np, video_boxes)]
video_high = [highlight_boxes(v, b) for v, b in zip(video_np, video_boxes)]
```

## Optical flow: regular
* On original video
* On detected boxes

```python
def run_optical_flow(video):
    frames = []
    # params for ShiTomasi corner detection
    feature_params = dict( maxCorners = 100,
                       qualityLevel = 0.3,
                       minDistance = 7,
                       blockSize = 7)
    # Parameters for lucas kanade optical flow
    lk_params = dict( winSize  = (15,15),
                  maxLevel = 2,
                  criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 10, 0.03))
    # Create some random colors
    color = np.random.randint(0,255,(100,3))
    # Take first frame and find corners in it
    old_frame = video[0]
    old_gray = cv2.cvtColor(old_frame, cv2.COLOR_BGR2GRAY)
    p0 = cv2.goodFeaturesToTrack(old_gray, mask = None, **feature_params)
    # Create a mask image for drawing purposes
    mask = np.zeros_like(old_frame)
    for i in range(1, len(video)):
        frame = video[i]
        frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        # calculate optical flow
        p1, st, err = cv2.calcOpticalFlowPyrLK(old_gray, frame_gray, p0, None, **lk_params)
        # Select good points
        good_new = p1[st==1]
        good_old = p0[st==1]
        # draw the tracks
        for i,(new,old) in enumerate(zip(good_new,good_old)):
            a,b = new.ravel()
            c,d = old.ravel()
            mask = cv2.line(mask, (a,b),(c,d), color[i].tolist(), 2)
            frame = cv2.circle(frame,(a,b),5,color[i].tolist(),-1)
        img = cv2.add(frame,mask)
        frames.append(img)
        # Now update the previous frame and previous points
        old_gray = frame_gray.copy()
        p0 = good_new.reshape(-1,1,2)
    return frames
```

```python
simple_flow = run_optical_flow(video_np)
simple_flow = [ cv2.cvtColor(v, cv2.COLOR_BGR2RGB) for v in simple_flow ]
show_images(simple_flow[(len(simple_flow) // 2) + 2 :(len(simple_flow) // 2  + 6)], cols=1)
```

```python
simple_flow_high = run_optical_flow(video_high)
show_images(simple_flow_high[(len(simple_flow_high) // 2) + 2 :(len(simple_flow_high) // 2  + 6)], cols=1)
```

## Dense optical flow

```python
def run_dense_optical_flow(video):
    frames=[]
    frame1 = video[0]
    prvs = cv2.cvtColor(frame1,cv2.COLOR_BGR2GRAY)
    hsv = np.zeros_like(frame1)
    hsv[...,1] = 255
    for i in range(1, len(video)):
        frame2 = video[i]
        nxt = cv2.cvtColor(frame2,cv2.COLOR_BGR2GRAY)
        flow = cv2.calcOpticalFlowFarneback(prvs,nxt, None, 0.5, 3, 15, 3, 5, 1.2, 0)
        mag, ang = cv2.cartToPolar(flow[...,0], flow[...,1])
        hsv[...,0] = ang*180/np.pi/2
        hsv[...,2] = cv2.normalize(mag,None,0,255,cv2.NORM_MINMAX)
        bgr = cv2.cvtColor(hsv,cv2.COLOR_HSV2BGR)
        frames.append(bgr)
        prvs = nxt
    return frames
```

```python
dense_flow = run_dense_optical_flow(video_np)
dense_flow = [ cv2.cvtColor(v, cv2.COLOR_BGR2RGB) for v in dense_flow ]
show_images(dense_flow[(len(dense_flow) // 2) + 2 :(len(dense_flow) // 2  + 6)], cols=1)
```

```python
dense_flow_high = run_dense_optical_flow(video_high)
show_images(dense_flow_high[(len(dense_flow_high) // 2) + 2 :(len(dense_flow_high) // 2  + 6)], cols=1)
```
