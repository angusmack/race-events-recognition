```python
import cv2
import numpy as np
from utils.video import * 
from utils.image import * 
import os
import pickle
import pypln as pp
from pypln.utils.pipeutils import infshuffle
from pipe import *
```

```python
# configuration variables
data_dir = 'videos'
classes = { 'collisions' : 1, 'not-collisions': 0 }
```

```python
def load_optical(filename): 
    return pickle.load(open(filename + '.pickle', 'rb'))


def calc_gradients(flow):
    def calc(flowi):
        old, new = zip(*flowi)
        return (np.array(new) - np.array(old)).T
    return [calc(flowi) for flowi in flow if flowi]


def to_polar(video):
    return [np.array(cv2.cartToPolar(frame[0], frame[1])).squeeze(axis=2) for frame in video]


def grad_to_hist(grad, bins, lower, upper, maxv=1):
    grad[grad > upper] = upper
    grad[grad < lower] = lower
    hist, _ =  np.histogram(grad, bins=bins, range=(lower, upper))
    hist = np.log(1 + hist) / maxv
    return hist


def video_to_hist(video, params):
    hists = []
    for frame in video:
        hists.append(np.array([
            grad_to_hist(frame[0], params[0]['bins'], params[0]['lower'], params[0]['upper'], params[0]['maxv']),
            grad_to_hist(frame[1], params[1]['bins'], params[1]['lower'], params[1]['upper'], params[1]['maxv'])
        ]).T)
    return np.array(hists)


def zero_pad(matrix, shape):
    matrix = np.array(matrix)
    if len(matrix.shape) != 3 or matrix.shape > shape:
        return np.zeros(shape)
    st = shape[0] - matrix.shape[0]
    col = shape[1] - matrix.shape[1]
    zs = shape[2] - matrix.shape[2]
    return np.pad(matrix, ((0, st), (0, col), (0, zs)), 'constant', constant_values=(0))
```

```python
## Calc basic statistics for normalization

shape = (190, 100, 2)

hist_params = [
    dict(
        bins=shape[1],
        lower=5,
        upper=240,
        maxv=1
    ),
    dict(
        bins=shape[1],
        lower=0,
        upper=6.29,
        maxv=1 
    ),
]

stream = (
    pp.get_datastream(data_dir, 'mp4', classes, split_filename='split.txt')
    | pp.apply('filename', 'rflows', load_optical)
    | pp.apply('rflows', 'rflows', calc_gradients)
    | pp.apply('rflows', 'rflows', to_polar)
    | pp.apply('rflows', 'rflows', lambda x: video_to_hist(x, hist_params))
    | pp.apply('rflows', 'rflows', lambda x: zero_pad(x, shape))
    | pp.filtersplit('T') 
    | select(lambda x: x['rflows']) 
    | as_list
)
maxes = np.max(stream, axis=(0,1,2))
hist_params[0]['maxv'] = maxes[0]
hist_params[1]['maxv'] = maxes[1]
del stream
```

```python
## Compute Optical Flow Descriptors


trainstream = (
    pp.get_datastream(data_dir, 'mp4', classes, split_filename='split.txt')
    | pp.apply('filename', 'rflows', load_optical)
    | pp.apply('rflows', 'rflows', calc_gradients)
    | pp.apply('rflows', 'rflows', to_polar)
    | pp.apply('rflows', 'rflows', lambda x: video_to_hist(x, hist_params))
    | pp.apply('rflows', 'rflows', lambda x: zero_pad(x, shape))
    | pp.filtersplit('T') 
    | infshuffle() 
    | pp.as_batch('rflows', 'class_id') 
)

valstream = (
    pp.get_datastream(data_dir, 'mp4', classes, split_filename='split.txt')
    | pp.apply('filename', 'rflows', load_optical)
    | pp.apply('rflows', 'rflows', calc_gradients)
    | pp.apply('rflows', 'rflows', to_polar)
    | pp.apply('rflows', 'rflows', lambda x: video_to_hist(x, hist_params))
    | pp.apply('rflows', 'rflows', lambda x: zero_pad(x, shape))
    | pp.filtersplit('V') 
    | infshuffle() 
    | pp.as_batch('rflows', 'class_id') 
)
```

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.AveragePooling2D((2, 2), input_shape=shape))
model.add(layers.Conv2D(64, (5,2), activation='relu', kernel_initializer='glorot_uniform'))
model.add(layers.Conv2D(32, (2,4), activation='relu', kernel_initializer='glorot_uniform'))
model.add(layers.MaxPooling2D(pool_size=(3, 3)))
model.add(layers.Dropout(0.3))

model.add(layers.Conv2D(32, (3,3), activation='relu', kernel_initializer='glorot_uniform'))
model.add(layers.Conv2D(32, (3,3), activation='relu', kernel_initializer='glorot_uniform'))
model.add(layers.Dropout(0.2))
model.add(layers.Flatten())

model.add(layers.Dense(100, activation='relu', kernel_initializer='glorot_uniform'))
model.add(layers.Dense(10, activation='relu', kernel_initializer='glorot_uniform'))
model.add(layers.Dense(1, activation='sigmoid', kernel_initializer='glorot_uniform'))

model.summary()
```

```python
from keras import optimizers
from keras import callbacks
import math


model.compile(loss='binary_crossentropy',
              optimizer=optimizers.Adam(lr=0.001),
              metrics=['acc'])

filepath = "weights-improvement-{epoch:02d}-{val_acc:.2f}.hdf5"

checkpoint =  callbacks.ModelCheckpoint(filepath, 
                          monitor='val_loss', 
                          verbose=0, 
                          save_best_only=True, 
                          save_weights_only=False, 
                          mode='auto', 
                          period=1)

history = model.fit_generator(
      trainstream,
      steps_per_epoch=100, 
      validation_data=valstream, 
      validation_steps = 15,
      epochs=20,
      callbacks = [checkpoint])
```
